#### Exercise 18. Normalize the variable of prices for each population and plot the 4 histograms in the same plot (you must use the subset obtained in Exercise 14) (★★★)

For the normalization method, you can use the one you consider; there is not a single correct answer to this question. Print the plot and write in the Markdown cell a brief analysis about the plot.

Hint: You can help yourself by reviewing the *multihist* demo of Matplotlib.

```python
southbelt_subset.loc[:,'normalized price'] = southbelt_subset['price']

plt.figure(figsize=(10,5))

for pop in southbelt_pops_list:
    mask = southbelt_subset['level5'] == pop
    pop_data = southbelt_subset.loc[mask, 'normalized price']
    if mask.any():
       max_price = pop_data.max()
       min_price = pop_data.min()
       southbelt_subset.loc[mask,'normalized price'] = (pop_data - min_price)/(max_price - min_price)
    plt.hist(southbelt_subset[southbelt_subset['level5']==pop]['normalized price'],bins=30, alpha=0.5, label=str(pop), histtype='step')
    plt.legend()
plt.show()
```

**Plot Analysis:** Comparing the normalized distributions of the different populations in the south belt, there is a clear concentration of lower price properties in Alcorcón and a big portion of the properties in Getafe (between 0.0 and 0.05).

Fuenlabrada shows a pronounced peak not very far from Alcorcón's and Getafe's (between 0.1 and 0.15). It shares in this area several data points with the other three populations, especially with Leganés that has its peak in that range, but at a much lower frequency and a distribution skewed to the right, where the rest of its values lay, on a relatively wide tail compared to the other distributions.

Getafe shows a second peak on its distribution, between 0.3 and 0.35 and standing out in the plot. This suggests that this population may have two different features (or groups of features) that determine on which of the regions around the peaks the price of the property belongs.

#### Exercise 21. Make a plot of the coordinates (latitude and longitude columns) of the south belt of Madrid by color of each population (you must use the subset obtained in Exercise 14) (★★★★)

Execute the following cell, and then start coding in the next one. You must implement a simple code that transforms the coordinates columns in a Python dictionary (add more information if needed) and then add it to the map

```python
from ipyleaflet import Map, basemaps, Marker, AwesomeIcon
from IPython.display import display

# Map centered on (60 degrees latitude and -2.2 degrees longitude)
# Latitude, longitude
map = Map(center = (40.4168, -3.7038), zoom = 10, min_zoom = 1, max_zoom = 20, basemap=basemaps.OpenStreetMap.Mapnik)
display(map)
```

```python
southbelt_subset.loc[:, 'latitude'] = southbelt_subset['latitude'].str.replace(',','.').astype('float')
southbelt_subset.loc[:, 'longitude'] = southbelt_subset['longitude'].str.replace(',','.').astype('float')
```

```python
map.layers = [map.layers[0]]

## HERE: plot the coordinates of the estates


## PUT HERE YOUR CODE:
coordinates_dict = {}

def get_marker_icon(pop):
    if pop == 'Fuenlabrada':
        return AwesomeIcon(icon='info-sign', marker_color ='blue', icon_color='blue')
    elif pop == 'Leganés':
        return AwesomeIcon(icon='info-sign', marker_color ='orange', icon_color='orange')
    elif pop == 'Getafe':
        return AwesomeIcon(icon='info-sign', marker_color ='green', icon_color='green')
    else:
        return AwesomeIcon(icon='info-sign', marker_color ='red', icon_color='red')

for idx, row in southbelt_subset.iterrows():
    if row['latitude'] != 0:
        coordinates_dict[int(row['id_realEstates'])] = (row['latitude'], row['longitude']), get_marker_icon(row['level5'])

for id, ((lat, lon), icon) in coordinates_dict.items():
    marker = Marker(location=(lat, lon), draggable=False, icon=icon)
    map.add_layer(marker)

display(map) 
```
